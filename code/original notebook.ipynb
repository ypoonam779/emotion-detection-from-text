# Same thing can be achieved by using single function.

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words('english'))
lemma=WordNetLemmatizer()

def clean_text(cl_text):
    cl_text = re.sub(r'http\S+', '', cl_text)                         #Removing url
    cl_text = re.sub('[^a-zA-Z]',' ',cl_text)                         #Removing numbers and punctuations
    cl_text = str(cl_text).lower()                                    #Converting all characters into lowercase
    cl_text = word_tokenize(cl_text)                                  #Tokenization
    cl_text = [item for item in cl_text if item not in stop_words]    #Removing stopwords
    cl_text = [lemma.lemmatize(word=w,pos='v')for w in cl_text]       #Lemmatization
    cl_text = ' '.join(cl_text)                                       #converting tokens to string
    return cl_text
emts['clean_text']=emts['Text'].apply(clean_text)
```

With the help of a single function we can achieve data cleaning. There is no need of writing a separate code.

# Feature Extraction

Feature extraction is a process of transforming data into features which is capable used for machine learning model. Typically, machine learning algorithms programmed with numeric. Hence, the text or word is then mapped into numerical vectors feature (a series of numbers). This process is called text vectorization or Bag of Words representation.

The Bag of Words (BoW) model is the simplest form of text representation in numbers. Like the term itself, we can represent a sentence as a bag of words vector (a string of numbers).


1. Basic BOW  CountVectorizer

1. Basic BOW  CountVectorizer

The CountVectorizer computes the frequency of occurrence of a word in a document. It converts the corpus of multiple sentences (say product reviews) into a matrix of reviews & words & fills it with frequency of each word in a sentence.

```python
from sklearn.feature_extraction.text import CountVectorizer

word_vectorizer = CountVectorizer(ngram_range=(1, 1))
# fit and transform on it the training features
word_vectorizer.fit(X)
vectors = word_vectorizer.transform(X)
X_vec = vectors.toarray()
#print(X_vec.shape)
print(word_vectorizer.get_feature_names())
```

Here first we are importing the countvectorizer feature extraction from the sklearn library. First creating a object of countvectorizer and in the parameter we have given ngram_range(1,1) means an n-gram is just a string of n words in a row. E.g. the sentence I am Groot contains the 2-grams I am and am Groot. The sentence is itself a 3-gram. Set the parameter ngram_range=(a,b) where a is the minimum and b is the maximum size of ngrams we want to include in our features. The default ngram_range is (1,1). Then we are fitting on the training data to learn and then transforming it on the training data to create a matrix. At last printing the features as Unigram and biagram.

```python
print(X_vec)
```

# BALANCING THE DATASET WITH RANDOM OVERSAMPLING

As the dataset is imbalanced to make it balanced we are making use randomoversampler method.
Random oversampling involves randomly duplicating examples from the minority class and adding them to the training dataset.

Examples from the training dataset are selected randomly with replacement. This means that examples from the minority class can be chosen and added to the new more balanced training dataset multiple times; they are selected from the original training dataset, added to the new training dataset, and then returned or replaced in the original dataset, allowing them to be selected again.

This technique can be effective for those machine learning algorithms that are affected by a skewed distribution and where multiple duplicate examples for a given class can influence the fit of the model. This might include algorithms that iteratively learn coefficients, like artificial neural networks that use stochastic gradient descent. It can also affect models that seek good splits of the data, such as support vector machines and decision trees.

Random oversampling can be implemented using the RandomOverSampler class.

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=777)
X_ROS, y_ROS = ros.fit_sample(X_vec, y)
X_train, x_test, Y_train, y_test = train_test_split(X_ROS,y_ROS,test_size=0.3,random_state=42)
```

From the sklearn we are importing the train_test split library. Mostly it is 70: 30 ratio 70% for training and 30% for testing and random_state=0 is defined as we keep running the cell the accuracy keep changing if random_state is not defined.

# RANDOMFOREST CLASSIFIER

Random forest is a type of supervised machine learning algorithm based on ensemble learning. Ensemble learning is a type of learning where we join different types of algorithms or same algorithm multiple times to form a more powerful prediction model. The random forest algorithm combines multiple algorithm of the same type i.e. multiple decision trees, resulting in a forest of trees, hence the name "Random Forest". The random forest algorithm can be used for both regression and classification tasks.

```python
from sklearn.ensemble import RandomForestClassifier

rbf = RandomForestClassifier(n_estimators = 50)
rbf.fit(X_train,Y_train)
print('RandomForest Classifer Created')
```

Here we are importing libraries required for building a Random Forest classifier from sklearn. Creating a object of randomforest classifier, n_estimators specifies the number of trees we want to include in building the random forest and then using the fit method we train the model using the training set.

Here we are predicting the response for the test set using the random forest model

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score

print(confusion_matrix(y_test, pred_rbf))
print(classification_report(y_test, pred_rbf))
print(accuracy_score(y_test, pred_rbf))
```

Here,We are importing the necessary libraries from scikit-learn  for performance evaluation of the model. We have confusion matrix, classification report and accuracy score.

1. Confusion matrix is a performance measurement for machine learning classification problem where output can be two or more classes. It is a table with 4 different combinations of predicted and actual values.


2. Accuracy is expressed as the proportion of correctly classified cases over all cases and is calculated according to formula:
                   =(+)/(+++)
3. Precision is expressed as the proportion of positive cases that are correctly recognized as positive over all cases classified as positive and is calculated according to the formula:
                   =/(+)
4. Recall is expressed as the proportion of positive cases that are correctly recognized as positive over all actual positive cases and is calculated according to the formula:
                 recall=/(+)
5. A measure that combines precision and recall is called the F1-measure and represents their weighted harmonic mean. It is calculated according to the formula:
                1=(2)/(+)

```python
#Get the confusion matrix and put it into a df
cm = confusion_matrix(y_test, pred_rbf) 
cm_df = pd.DataFrame(cm,
                     index = ['anger','fear','joy', 'nuetral','sadness'], 
                     columns = ['anger','fear','joy', 'nuetral','sadness'])

#Plot the heatmap
plt.figure(figsize=(12, 8))

sns.heatmap(cm_df, 
            center=0, 
            cmap=sns.diverging_palette(220, 15, as_cmap=True), 
            annot=True, 
            fmt='g')
all_sample_title = 'Accuracy Score: {0}'.format(rbf.score(x_test, y_test))
plt.title(all_sample_title, size = 15)
plt.ylabel('True label', fontsize = 13)
plt.xlabel('Predicted label', fontsize = 13)
plt.show()
```
